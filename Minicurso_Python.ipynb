# **4. Definindo Ponto e Reta no Python**

No contexto de Geometria Analítica, um ponto é representado por suas coordenadas $(x, y)$ em um plano bidimensional, ou $(x, y, z)$ em um espaço tridimensional. No Python, podem ser representados como uma tupla ou lista.
* `Tupla` é uma estrutura de dados leve que possui um número específico e uma sequência de valores.
* `Lista` é um tipo de estrutura de dados que permite armazenar uma coleção de itens em uma única variável. Diferente da tupla, as listas são mutáveis, o que significa que podem ser modificadas após serem criadas.


Uma reta no Python pode ser definida por sua equação geral:
$$Ax + By + C = 0.$$

```python
def equacao_reta(A, B, C):
    return f"{A}x + {B}y + {C} = 0"
```

# Exemplo 1

Desenvolva um código em Python para representar a equação geral de uma reta $Ax + By + C = 0$, onde $A = 3$, $B = -4$ e $C = 5$. Com esses valores, deseja-se definir a reta.

Para obter a equação geral da reta que passa por dois pontos
$(x_1,y_1)$ e $(x_2,y_2)$, podemos seguir os dois procedimentos listados a seguir:
1. Calcular os coeficientes $A$, $B$ e $C$ da equação $Ax + By + C = 0$, onde:
* $A = y_2 - y_1$
* $B = x_1 - x_2$
* $C = x_2y_1 - x_1y_2$

2. Outra forma de obter a equação de uma reta que passa por dois pontos é usando a **forma reduzida da equação da reta**. A qual é dada por: $$y = m(x - x_1) +y_1,$$
onde $(x_1, y_1)$ é um ponto conhecido na reta, e $m$ representa a inclinação ou declividade da reta.
A declividade/inclinação pode ser calculada pela razão entre a variação em $y$ e a variação em $x$ entre dois pontos dados da reta. Portanto, conhecendo dois pontos $(x_1, y_1)$ e $(x_2, y_2)$, temos: $$m = \frac{y_2 - y_1}{x_2 - x_1}.$$
Podemos, ainda, considerar o ponto $(x_2, y_2)$ e escrever a equação da reta como: $$ y = m(x - x_2) +y_2.$$

```python
def equacao_geral_reta(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    # Calcula os coeficientes A, B e C
    A = y2 - y1
    B = x1 - x2
    C = x2 * y1 - x1 * y2
    # Retorna a equação geral na forma Ax + By + C = 0
    return f"{A}x + {B}y + {C} = 0"
```

# **5. Distância Entre um Ponto e uma Reta**

Para calcular a distância entre um ponto $(x_0,y_0)$ e uma reta dada pela equação geral $Ax + By + C = 0$, usamos a fórmula:
$$d = \frac{| Ax_0 + By_0 + C |}{\sqrt{A^2+B^2}}$$

### Note que no cálculo da distância $d$ temos uma raiz quadrada e o módulo (valor absoluto). Para isso, utilizaremos duas funções das bibliotecas já vistas.
* Tanto na biblioteca `numpy` quanto na `math`, o valor absoluto é obtido pela função `abs()`.
* No cálculo da raiz quadrada, utilizamos o nome ou apelido da biblioteca antes da função `sqrt`.

Exemplo: se formos usar a biblioteca `math`, escrevemos `math.sqrt()` e se usarmos a `numpy`, chamamos a funcão escrevendo `np.sqrt`.

```python
def distancia_ponto_reta(A, B, C, x0, y0):
    distancia = abs(A * x0 + B * y0 + C) / math.sqrt(A**2 + B**2)
    return distancia
```

# **6. Distância Entre Retas Paralelas**
O cálculo da distância entre duas retas paralelas é feito utilizando a fórmula: $$d(r,s) = \frac{|C_2 - C_1|}{\sqrt{A^2 + B^2}}$$
1. As equações das duas retas paralelas são
dadas na forma geral: $$ r: A_1x + B_1y + C_1 = 0$$ $$s: A_2x + B_2y + C_2 = 0$$
No caso de retas paralelas, os coeficientes $A_1$ e $B_1$ serão iguais aos de $A_2$ e $B_2$ devido a serem paralelas, isto é, por terem a mesma inclinação.
2. $|C_2 - C_1|$ representa o valor absoluto da diferença entre as constantes das duas equações. Lembre-se que $C_1$ e $C_2$ definem a posição das retas no plano.
3. $\sqrt{A^2+B^2}$ normaliza a distância, levando em conta a inclinação das retas. Isso ajusta a medida para garantir que a distância calculada seja perpendicular às duas retas.


```python
def distancia_entre_retas_paralelas(A, B, C1, C2):
    # Calcula a distância entre duas retas paralelas
    distancia = abs(C2 - C1) / math.sqrt(A**2 + B**2)
    return distancia
```
